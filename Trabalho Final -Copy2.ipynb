# Retirada de Outliers

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    print(fence_low)
    print(fence_high)
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

# Função de validação cruzada, intervalo de confiânça e rodar melhor modelo

```python
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support
import numpy as np
import copy
from sklearn.model_selection import KFold
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
```

```python
import copy
import numpy as np
from sklearn.metrics import precision_recall_fscore_support, classification_report
from scipy.stats import t

def MelhorModelo(modelo):
    melhor = 0
    fold = 1
    labels = ['0', '1']
    labels2 = [0, 1]

    precision_valid = [0, 0]
    recall_valid = [0, 0]
    fscore_valid = [0, 0]

    accuracies = []

    for train_index, valid_index in kf.split(X_train):
        x_train = X_train.iloc[train_index].loc[:]
        y_train = y_train2.iloc[train_index]    

        x_valid = X_train.iloc[valid_index].loc[:]
        y_valid = y_train2.iloc[valid_index]

        modelo.fit(x_train, y_train)
        predito = modelo.predict(x_valid)

        # Calculating the metrics
        precision, recall, fscore, support = precision_recall_fscore_support(y_valid, predito, average=None)   
        precision_valid = np.add(precision_valid, precision)
        recall_valid = np.add(recall_valid, recall)
        fscore_valid = np.add(fscore_valid, fscore)     

        acuracia = np.mean(y_valid == predito)
        accuracies.append(acuracia)

        if acuracia > melhor:          
            melhor = acuracia
            best_model = copy.deepcopy(modelo)
        fold += 1

    p = best_model.predict(X_test)
    classification_rep = classification_report(y_test, p)
    
    mean_accuracy = np.mean(accuracies)
    std_dev = np.std(accuracies)
    confidence_level = 0.95  
    n = len(accuracies)
    t_value = t.ppf((1 + confidence_level) / 2, n - 1)
    margin_error = t_value * std_dev / np.sqrt(n)
    lower_bound = mean_accuracy - margin_error
    upper_bound = mean_accuracy + margin_error
    accuracy_std = np.std(accuracies)



    print(classification_rep, (lower_bound, upper_bound))
    print(f"Desvio de {accuracy_std}")
    

```

# Função para realizar o teste T

```python
from scipy.stats import ttest_ind

def teste_t(model1_mean, model1_std, model2_mean, model2_std):
    _, p_value = ttest_ind(model1_mean, model2_mean, equal_var=False)
    mean_diff = np.mean(model1_mean) - np.mean(model2_mean)

    if p_value < 0.05:
        result = "Diferença significante"
    else:
        result = "Sem diferença significante"

    return result, p_value, mean_diff
```
